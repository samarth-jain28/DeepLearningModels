```python
import math, copy
import numpy as np
import matplotlib.pyplot as plt
```

```python

def fn(x, w, b):
    return w*x + b
```

```python
def compute_cost(x, y, w, b):
    cost = 0
    m = len(x)
    for i in range(m):
        cost += (fn(x[i], w, b) - y[i])**2
    return cost/(2*m)
```

```python
def compute_gradient(x, y, w, b):
    m = len(x)
    der_w = 0
    der_b = 0
    for i in range(m):
        der_w += (fn(x[i], w, b)-y[i]) * x[i]
        der_b += (fn(x[i], w, b)-y[i]) 
    return der_w/m, der_b/m

def gradient_descent(x, y, w, b, alpha, itr):
    costs = []
    it = []
    for i in range(itr):
        tmpw, tmpb = compute_gradient(x, y, w, b)
        w -= (tmpw*alpha)
        b -= (tmpb*alpha)
        tcost = compute_cost(x, y, w, b)
        costs.append(tcost)
        it.append(i+1)
    
    return w, b, [i, costs]
    
```

```python
# assigning random weight and bias in w and b parameters. 
x = [1.0, 2.0]
y = [300, 500]
b = -100
w = 100
cost = compute_cost(x,y, w, b)  # computing cost for random w & b
print(cost)

line = np.arange(-1000,1000,1)  # checking line for our function for random w & b
line = [fn(i, w, b) for i in range(10) ]
plt.plot(line)
plt.scatter(x,y)
plt.show()
```

```python
# checking cost for different values of w i.e weights
cost_list = []
for i in range(-10, 20, 1):
    cost_list.append([i, compute_cost(x, y, i, b)])

print(cost_list)
```

```python
# Call gradient descent and implement automatic learning
w = -100
b = 100
alpha = 1.0e-4
itr = 10000
w, b, costs = gradient_descent(x, y, w, b, alpha, itr)

print(w, b)

plt.plot(costs[1])
plt.xlabel("Iteration")
plt.ylabel("Costs")
plt.show
```
